```python
# Import the dependencies:
import pandas as pd
from dotenv import load_dotenv
import os
import requests
import alpaca_trade_api as tradeapi

# Load the env file
load_dotenv()
```

```python
# Create the API and secret key variable object:
alpaca_api_key = os.getenv('APCA_API_KEY_ID')
alpaca_secret_key = os.getenv('APCA_API_SECRET_KEY')

# Create the tradeapi.REST object using the API & secret keys, and setting the api_version to 'v2':
alpaca = tradeapi.REST(
    alpaca_api_key,
    alpaca_secret_key,
    api_version='v2'
)

# Get the tickers for the stocks that you'll be analyzing:
tickers = ['AMZN', 'TSLA']

# Set the Timeframe to 1Day
timeframe = '1Day'

# Set the start and end dates
start_date = pd.Timestamp('2018-12-20', tz='America/New_York').isoformat()
end_date = pd.Timestamp('2023-12-20', tz='America/New_York').isoformat()


# Create the `anaya_stocks_df` DataFrame using the `get_bars()` and .df functions:
anaya_stocks_df = alpaca.get_bars(
    tickers,
    timeframe,
    start=start_date,
    end=end_date
).df

# Review the DataFrame:
display(anaya_stocks_df.head())
display(anaya_stocks_df.tail())
```
